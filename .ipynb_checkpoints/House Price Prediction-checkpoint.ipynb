```python
import pandas as pd 


train_data = pd.read_csv("C:/Users/Michael/Downloads/house price dataset.csv")

test_data = pd.read_csv("C:/Users/Michael/Downloads/house price dataset.csv")

train_data.head()
```

```python
# Check the shape of the dataset
print(f"Training data shape: {train_data.shape}")
print(f"Test data shape: {test_data.shape}")

# Get data types and non-null counts
train_data.info()
```

```python
# Summary statistics for numerical features
train_data.describe()
```

```python
# Total missing values per column
missing_values = train_data.isnull().sum()
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)
print(missing_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(train_data['price'], kde=True)
plt.title('Distribution of prices')
plt.xlabel('price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Percentage of missing values
percent_missing = train_data.isnull().sum() * 100 / len(train_data)
missing_data = pd.DataFrame({'Missing Percentage': percent_missing})
missing_data = missing_data[missing_data['Missing Percentage'] > 0]
print(missing_data.sort_values(by='Missing Percentage', ascending=False))
```

```python
import numpy as np
# Fill missing numerical values with median
numerical_features = train_data.select_dtypes(include=[np.number]).columns
for col in numerical_features:
    median = train_data[col].median()
    train_data[col].fillna(median, inplace=True)
    test_data[col].fillna(median, inplace=True)

# Fill missing categorical values with mode
categorical_features = train_data.select_dtypes(include=['object']).columns
for col in categorical_features:
    mode = train_data[col].mode()[0]
    train_data[col].fillna(mode, inplace=True)
    test_data[col].fillna(mode, inplace=True)
```

```python
# Combine train and test data for consistent encoding
combined = pd.concat([train_data.drop('price', axis=1), test_data], axis=0)

# One-Hot Encoding
combined_encoded = pd.get_dummies(combined)

# Split back into train and test
train_encoded = combined_encoded.iloc[:train_data.shape[0], :]
test_encoded = combined_encoded.iloc[train_data.shape[0]:, :]

# Ensure no data leakage
X = train_encoded
y = train_data['price']
```

```python
X = X.copy()  # If X was a slice of another DataFrame

# Method 1: Fill NaN values with column mean
X = X.fillna(X.mean())

# Method 2: Replace infinite values with NaN and then fill them
# First replace inf values with NaN
X = X.replace([np.inf, -np.inf], np.nan)
# Then fill NaN values with column mean
X = X.fillna(X.mean())

# Alternative one-liner combining both operations
X = X.replace([np.inf, -np.inf], np.nan).fillna(X.mean())
```

```python
# Check for NaN values
print(X.isnull().sum())

# Check for infinite values
print(np.isinf(X).sum())
```

```python
# Replace NaN values with the mean of each column
X.fillna(X.mean(), inplace=True)

# Replace infinite values with NaN, and then handle them
X.replace([np.inf, -np.inf], np.nan, inplace=True)
X.fillna(X.mean(), inplace=True)
```
